```r
# Selects the row that meets the condition
bike_df  %>% 
    filter(season == 2, workingday == 1)
```

> We want to find out the records of rented bikes on a working day during summer from 2011-2012. To get the records, we select the row  using `filter()`, based on the condition that they are recorded during summer, and it is a working day. This yields 3,066 rows with all of the information for each of them.

```r
# Order the rows in ascending order based on humidity.
bike_df %>% 
    arrange(hum)
```

> We want to find out if more are renting bikes when the humidity is high. Just by relying on the result from arranging the rows in ascending order, it seems that there are more casual and registered users renting a bike when the humidity is high, with registered users renting more bikes than casual users.

```r
# The following lines of code find and return the topmost 40 rows based on the total number of rented bikes.
bike_df  %>% 
    filter(season == 2, workingday == 1) %>% 
    arrange(hum) %>% 
    top_n(40, cnt) 
```

> In this case, we combined the previous lines of code to find the topmost 40 records with the high number of bikes rented during summer on a working day using `top_n()`. From the resulting tibble, there seems to be a high volume of bikes rented in the summer of 2012. Furthermore, the common rental time is around late afternoon to evening, and, interestingly, there are some records where the humidity is relatively low, which contradicts the findings earlier.

```r
# Our own implementation of top_n() that returns nth top rows based on the variable specified.
new_topn <- function(df, var, index) {
    result <- df[order(var, decreasing = TRUE),]
    return(result[1:index,])
}

new_df <- bike_df %>% 
    filter(season == 2, workingday == 1) %>% 
    arrange(hum) 
new_df %>% 
    new_topn(var = new_df$cnt, index = 40)
```

> We implemented our `top_n()` using `order()` which sorts the row in descending order based on the variable specified, which in this case is the cnt variable, and then selects the top nth rows by the value stated. Using our `top_n()`, we could show the same information from previous lines of code that uses the original `top_n()`.

<h2>TASK (D) GROUPS AND NOT</h2>

```r
# Modify the bike_df by creating and adding new columns of proportions.
bike_df  %>% 
    mutate(casual_prop = round(casual / cnt, 2), registered_prop = round(registered / cnt, 2)) 
```

> We want to find the proportion of rented bikes for each observation and include it in **bike_df**. To find the proportion of rented bikes for each user type for every observation, we take the number of rented bikes for each user type and then divide it by the total rented bikes for each observation. Then round them off using `round()` and save them by assigning them to their column name (**casual_prop** and **registered_prop**). Lastly, we use `mutate()` to add these columns to **bike_df**.

```r
# Creates a new dataframe containing the mean of the humidity of the whole bike_df
bike_df %>% 
    summarise(hummean = mean(hum))
```

>  To summarise the bike_df based on the mean humidity, we use `summarise()` and `mean()` to reduce the dataframe to the resulting mean humidity. This allows us to find the average humidity value of the dataframe, which gives us the typical humidity value where users rented a bike.

```r
# Group the rows by season and sum up the total rented bikes by casual users per season.
bike_df  %>% 
    group_by(season)  %>% 
    tally(casual) 
```

> We want to determine the volume of rented bikes by casual users per season from 2011 to 2012. To get this information, we initially group the rows by season using `group_by()` and then use `tally()` to sum up the total number of rented bikes by casual users per season. From the resulting tibble, many casual users rent bikes during summer and fall while there are significantly less in spring.

```r
# Group the rows by season and summarise the dataframe through five number summary based on the total rented bikes per season.
bike_df  %>% 
    group_by(season)  %>% 
    summarise(mean = mean(cnt),
              min = min(cnt),
              q1 = quantile(cnt, 0.25),
              median = median(cnt),
              q3 = quantile(cnt, 0.75),
              max = max(cnt))
    
```

> Using the `group_by()` and `summarise()`, we constructed a five-number summary table which gives us a rough idea about the distribution of the data depending on the season. They exhibit positive skewness and do not follow a symmetric shape as the mean is greater than the median.

<h2>TASK (E) WIDE, LONG, TIDY</h2>

```r
# Reshapes the bike_df into a long data format.
(long_ver <- bike_df %>% 
    gather(key = day_type,
           value = day_value, c(holiday, workingday)) %>% 
    na.omit())
```

> To convert the **bike_df** to a long data format, we use `gather()` where it creates a key column (**day_type**) that takes the **holiday** and **workingday** that are formerly columns as its values. It also makes a value column (**day_value**)  that takes the values of the two former columns as its value. It doubles the number of rows as each row corresponds to one observation per value of the key column. We added `na.omit()` to eliminate any rows with missing data.

```r
# Reshapes the long data version of bike_df to wide data format (back to its original shape).
long_ver %>% 
    spread(key = day_type,
           value = day_value) %>% 
    na.omit()
```

> To convert the **bike_df** to its original form, we use `spread()` where it takes the values of created **day_type** and turns them back as column names. Then, it fills the columns with the values from the **day_value** column. However, it is noticeable that the columns are not in their original position, and the number of rows returns to its original state. We use `na.omit()` again to remove existing rows with NA.

```r
# bike_df tidy version
(tidy_ver <- as_tibble(bike_df))
```

```r
# Takes the bike_df and modify it by changing the column type of holiday and season variables to factor type to represent each 
# of their levels. Save this modified bike_df as temp_df for plot creation.
temp_df <- bike_df  %>% 
    mutate(holiday = factor(holiday),
           season = factor(season, labels = c("Spring", "Summer", "Fall", "Winter")))
temp_df %>% glimpse()
```

```r
# Creates a boxplot that compares the normalized temperature of day types, holiday and non-holiday per season.
# Then print out the plot.
(temp_plot <- temp_df %>% ggplot(aes( x = holiday, y = temp, color = holiday)) + 
    geom_boxplot() +
    facet_wrap(~season) +
    theme(legend.position="right") +
    scale_color_discrete(name = "Day Type", labels = c("Holiday", "Working Day")) +
    theme(strip.text.x = element_text(size = 13, color = "black", face = "bold.italic")) +
    scale_x_discrete(labels = c("Holiday", "Working Day")) +
    xlab("Holiday") + ylab("Normalized Temperature (C)") + 
    labs(title = "Temperature comparison for all season and day type (2011-2012)"))
```

```r
# A grouped dataframe that contains the columns season, hour, and total number of bikes rented by registered users. 
(challengeone_df <- bike_df %>% 
    mutate(season = factor(season)) %>% 
    filter(hr >= 6, hr <= 10) %>% 
    group_by(season, hr) %>% 
    tally(registered))
```

> To test the hypothesis of the CEO, we first created a grouped dataframe by mutating the bike_df, where we converted the season into a factor type. We then filter it by the period, which the CEO thinks is one of the peak times. After that, we group the data by season and hour to get the total bikes rented depending on them. 

```r
# A jitter plot that shows the morning peak rental period by season
challengeone_df %>% 
    ggplot(aes(x = hr, y = n, color = season)) + # creates empty axis
    geom_point(size = 5, alpha = 0.8) + # plots point in the empty axis
    theme(legend.position="right") + # change the legend position
    scale_color_discrete(name = "Season", labels = c("Spring", "Summer", "Fall", "Winter")) + # change the legend title and values
    xlab("Time of the Day (hour)") + ylab("Counts of Bike Rented by Registered Users") + # add labels to x and y axis
    labs(title = "Total of rented bikes from 6am-10am per season (2011-2012)") + # add title to the plot
    ylim(0, 80000) # change the limit range of y-axis
```

> We use the new dataframe (**challengeone_df**) to create the jitter plot to show and prove that one of the peaks is between 6 am and 10 am. We use `ggplot()` and its aesthetics to create a jitter plot that one of the peak rental times is indeed between 6 am and 10 am.

```r
# A grouped dataframe that contains the columns season, hour, weathersit, and total number of bikes rented by registered users.  
(next_morning <- bike_df %>% 
    mutate(season = factor(season, labels = c("Spring", "Summer", "Fall", "Winter")),
           weathersit = factor(weathersit)) %>% 
    filter(hr >= 6, hr <= 10) %>% 
    group_by(season, hr, weathersit) %>% 
    tally(registered))
```

```r
# A jitter plot that shows the amount of bikes rented by registered between 6 am and 10 am depending on season and previous
# weather situation.
next_morning %>% 
    ggplot(aes(x = hr, y = n, color = weathersit)) + # creates empty axis
    geom_point(size = 3, alpha = 0.8) + # plots point in the empty axis
    facet_wrap(~season) + # creates a panel based on season
    scale_color_discrete(name = "Weather") + # change the legend title
    theme(strip.text.x = element_text(size = 13, color = "black", face = "bold.italic")) + # change the panel text size, color, and style
    xlab("Time of the Day (hour)") + ylab("Counts of Bike Rented by Registered Users") + # add label to x and y axis
    labs(title = "Rented bikes from 6am-10am based on weather per season") # add title to the plot
```

> In order to test the second hypothesis of the CEO, created a new dataframe that contains season, hour, weather situation, and the total number of bikes rented by registered users columns. We made it by initially mutating the season and weathersit column (converting them as a factor). We then filter the rows to get a subset of observations recorded between 6 am, and 10 am. After that, we group it by season, hour, and weather situation. Lastly, we sum up the number of bikes rented by registered users.

> We use the created dataframe to create the jitter plot shown above. The jitter plot displays the volume of bikes rented by registered users per season with knowledge about the weather situation from the previous afternoon and depending on it. For example, weather situations 3 and 4 can be classified as poor weather conditions, and it shows if the previous afternoon has these situations, then there will be fewer or no registered users who will rent a bike the following day between 6 am and 10 am. This is quite the opposite for situations 1 and 2, with situation 1 always guaranteeing a registered user will rent a bike the following day. There is also a noticeable seasonal effect.
